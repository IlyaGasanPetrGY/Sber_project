```python
ga_session_duplicate['list_time'] = ga_session_duplicate['visit_time'].apply(lambda x: x.split(':'))
```

```python
def getting_time_seconds(hor, mine, sec):
    print(int(hore)*60*60 + int(mine)*60 + int(sec))
    return int(hore)*60*60 + int(mine)*60 + int(sec) 
    
ga_session_duplicate['time_in_seconds'] = ga_session_duplicate['list_time'].apply(lambda x: int(x[0])*60*60 + int(x[1])*60 + int(x[2]))
```
