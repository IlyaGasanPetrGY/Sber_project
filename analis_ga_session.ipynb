#### Отчистка данных от неинформативных колонок

```python
def info_null(df:pd.DataFrame) -> list:
        columns_drop = []
        for i in df.columns:
            procent_null = df[i].isna().sum() / df.shape[0] * 100
            print(f'{i} - {procent_null:.2f}')
            if procent_null > 25:
                  columns_drop.append(i)
        return columns_drop

ga_session_drop = info_null(ga_session_duplicate)
print('/////////////////////////////')
hist_drop = info_null(hist_df_duplicat)
```

```python
clear_df_ga = ga_session_duplicate.drop(columns=ga_session_drop)
clear_df_hist = hist_df.drop(columns=hist_drop)
```
